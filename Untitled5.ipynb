Monte Carlo Simulation
Example For AAPL Stock.

```python
S0 = data['AAPL'].iloc[-1]   # current price
mu = log_returns['AAPL'].mean()
sigma = log_returns['AAPL'].std()

T = 252      # trading days in 1 year
simulations = 1000

price_paths = np.zeros((T, simulations))
```

Run Simulation:

```python
for i in range(simulations):
    prices = [S0]

    for t in range(T):
        shock = np.random.normal(mu, sigma)
        price = prices[-1] * np.exp(shock)
        prices.append(price)

    price_paths[:, i] = prices[1:]
```

Plot Simulations:

```python
plt.figure(figsize=(12,6))
plt.plot(price_paths)
plt.title("Monte Carlo Simulation for AAPL Price")
plt.show()
```
